### Movies per country

```python
d2 = data[['country','kind']].value_counts()
d2
```

```python
data_mean = data[["kind", "genre",'country']].groupby(["country"]).agg({
    'kind':'count','genre':'count'})
#preparing new DF that will have aggregated data : vehicles_mean
data_mean.plot.barh(color = {"purple","grey"}) 
```

```python
#Movies per country
df = data[data['country']!='United States']
countries = df.groupby('country')['kind'].value_counts().unstack()
countries.plot.bar(stacked=True)
```

```python
df1 = data[data['country']=='United States']
colorgrp = df1.groupby('genre')['kind'].value_counts().unstack()
colorgrp.plot.bar(stacked=True)
```

```python
data2 = data
data2["kind"] = np.where(data2['kind']=='video game', 'tv short + video game', data2['kind'])
data2["kind"] = np.where(data2['kind']=='tv short', 'tv short + video game', data2['kind'])

pie = data2.groupby(['kind']).agg('count').reset_index()
pie
```
